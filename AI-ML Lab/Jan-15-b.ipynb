```python
def median(arr):
    count = len(arr)
    mid = count // 2     #11 -> 5 = ok; 10 - > 5, (4th+5th)/2

    if count % 2 == 0:
        return (arr[mid] + arr[mid-1]) / 2 
    else:
        return arr[mid]
```
